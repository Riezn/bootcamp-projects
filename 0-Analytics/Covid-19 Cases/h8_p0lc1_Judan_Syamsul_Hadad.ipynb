## Soal Nomor 1

1. Menggunakan Pandas query, ambil data kasus kumulatif Covid-19 tanggal 20 Januari 2022 untuk negara-negara dengan latitude di bawah 0. Negara mana yang paling tinggi jumlah kumulatif kasus Covidnya per tanggal tersebut? Jika bukan Indonesia, bandingkan jumlah kumulatif kasus negara tersebut dengan Indonesia per tanggal yang sama!

### Import Library

```python
import numpy as np
import pandas as pd
```

### Data Loading

```python
df = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv") #read csv dari link
```

Lihat isi dataset

```python
df.head() #cek lima baris teratas
```

### Answer

Data kasus Covid-19 tanggal 20 Januari 2022 di negara dengan Latitude di bawah 0

```python
df[
    (df['Lat']<0) #query untuk baris dengan nilai Lat di bawah 0
][['Country/Region','Lat','1/20/22']] #menampilkan kolom Country/Region, Lat, dan tanggal 20 Januari 2022
```

Sekarang cek jumlah kasus kumulatif terbanyak di dunia pada query di atas

```python
df[
    (df['Lat']<0) #query untuk baris dengan nilai Lat di bawah 0
].groupby(['Country/Region'])['1/20/22'].max().max() #grouping berdasarkan negara agar nilai kasus max per negara dapat diketahui, lalu mengambil nilai max dari semua negara
```

Kasus kumulatif terbanyak pada 20 Januari 2022 sebanyak 23.595.178 kasus. Sekarang cek negara mana yang memiliki kasus terbanyak tersebut.

```python
df[
    (df['1/20/22']==23595178) #query negara dengan kasus 23595178 pada 20 Januari 2022 
]
```

Negara dengan kasus terbanyak tersebut adalah Brazil. Bandingkan dengan Indonesia.

```python
df[
    (df['Country/Region']=='Indonesia') #query di Indonesia
]['1/20/22'] #menampilkan kolom 20 Januari 2022
```

Indonesia memiliki kasus kumulatif sebanyak 4.277.644 pada tanggal yang sama. Kira-kira 1/6 dari Brazil.

## Soal Nomor 2

2. Buatlah fungsi untuk menghitung jumlah kasus harian Covid-19! (Hint: Kamu bisa menggunakan .diff() untuk menghitung perbedaan tiap baris data/menghitung jumlah kasus harian dan .transpose() untuk mengubah kolom menjadi baris atau sebaliknya)

### Import Library dan Data Loading

Library dan dataset yang digunakan masih mengikuti Soal Nomor 1 sehingga tidak di-load ulang.

### Answer

Sebelum membuat fungsi, kita sederhanakan dulu dataset dengan membuat copy dataset, mengubah label menjadi negara, drop beberapa kolom di depan, lalu transpose dataset tersebut.

```python
df_harian = df.copy() #copy dataset ke variabel baru
```

```python
df_harian.head() #cek lima baris teratas
```

Mengubah label menggunakan 'Country/Region'

```python
df_harian.set_axis(df_harian['Country/Region'],inplace=True) #mengubah label axis=0 atau baris menggunakan isi kolom Country/Region
```

```python
df_harian.head() #cek lima baris teratas
```

Drop beberapa kolom

```python
df_harian.drop(columns=['Province/State','Country/Region','Lat','Long'],inplace=True) #drop kolom yang tidak dibutuhkan untuk perhitungan kasus harian
```

```python
df_harian.head() #cek lima baris teratas
```

Transpose dataset

```python
df_harian = df_harian.T #transpose dataset agar negara menjadi kolom dan tanggal menjadi baris
```

```python
df_harian.head() #cek lima baris teratas
```

Pembuatan fungsi

```python
def kasus_harian(dataframe_input): #deklarasi fungsi dengan nama fungsi kasus_harian, input fungsi berupa dataframe
    dataframe_output = dataframe_input.diff()  #selisih nilai harian dihitung menggunakan .diff() lalu disimpan sebagai variabel dataframe_output
    return dataframe_output #fungsi mengembalikan dataframe_output 
```

Panggil fungsi dengan dataset yang sudah dibersihkan tadi dan simpan sebagai df_harian yang baru

```python
df_harian = kasus_harian(df_harian) #panggil fungsi dan disimpan sebagai df_harian
df_harian.head() #cek lima baris teratas
```

Nilai NaN dapat kita ubah ke 0 karena pada 22 Januari 2020 belum ada kasus sama sekali.

```python
df_harian.fillna(0,inplace=True) #mengubah NaN menjadi 0
```

```python
df_harian.head() #cek lima baris teratas
```

## Soal Nomor 4

4. Gunakan for loop dan conditional if untuk membuat sebuah list yang berisikan jumlah kasus harian Covid-19 dengan ketentuan hanya mengambil data Indonesia saja dan jumlah kasus di bawah 10000. Berapa jumlah hari dengan kasus di bawah 10000?

### Import Library dan Data Loading

Library dan dataset yang digunakan masih mengikuti Soal Nomor 2 dan 3 sehingga tidak di-load ulang.

### Answer

Buat for loop terlebih dahulu untuk mengumpulkan elemen dengan nilai di bawah 10.000 dari dataset df_harian

```python
kasus_under_10k = [] #buat list kosong untuk menyimpan output loop
i=0 #variabel untuk indexing
for kasus in df_harian['Indonesia']: #loop untuk tiap nilai di dalam df_harian kolom Indonesia
    if df_harian.iloc[i]['Indonesia']<10000: #condition: Jika kasus di bawah 10000 di Indonesia, maka 
        kasus_under_10k.append(kasus) #masukkan nilai kasus ke variabel kasus_under_10k
    i+=1 #index +1
```

Output loop disimpan dalam variabel kasus_under_10k. Sekarang kita hanya perlu menghitung jumlah elemen dalam list ini untuk mendapatkan jumlah hari.

```python
len(kasus_under_10k) #cek panjang list
```

Terdapat 633 hari dengan jumlah kasus di bawah 10.000.
